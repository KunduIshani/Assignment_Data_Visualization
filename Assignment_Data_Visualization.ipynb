```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Seed for reproducibility
np.random.seed(20)

# Create the dataset
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Pivot the DataFrame to create a matrix for the heatmap
heatmap_data = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')

# Create a heatmap using Plotly
fig = go.Figure(data=go.Heatmap(
    z=heatmap_data.values,
    x=heatmap_data.columns,
    y=heatmap_data.index,
    colorscale='Viridis'
))

# Update layout for better readability
fig.update_layout(
    title='Sales Variation Across Different Months and Days',
    xaxis_title='Month',
    yaxis_title='Day'
)

# Show the plot
fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Generate the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calculate the z values based on the function
z = np.sin(np.sqrt(x**2 + y**2))

# Flatten the arrays and create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout for better readability
fig.update_layout(
    title='3D Surface Plot of the Function',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
```

```python

     fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
     counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Prepare the output notebook (for Jupyter)
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Define a color palette
colors = ['#f28e2b', '#e15759', '#76b7b2', '#59a14f']

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', toolbar_location=None, tools="")

# Add bars to the plot
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits", line_color='white', fill_color=factor_cmap('fruits', palette=colors, factors=fruits))

# Customize the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the results
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Prepare the output notebook (for Jupyter)
output_notebook()

# Given data
data_hist = np.random.randn(1000)  # Generate 1000 random data points
hist, edges = np.histogram(data_hist, bins=30)  # Compute histogram

# Create a new plot with a title and axis labels
p = figure(title="Histogram of Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph to create the histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.7)

# Show the results
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
     data_heatmap = np.random.rand(10, 10)
     x = np.linspace(0, 1, 10)
     y = np.linspace(0, 1, 10)
     xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar, ColumnDataSource
from bokeh.transform import transform
import numpy as np

# Prepare the output notebook (for Jupyter)
output_notebook()

# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the grid and data for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Define a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_flat.min(), high=data_flat.max())

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, data=data_flat,
                                    left=x_flat - 0.05, right=x_flat + 0.05,
                                    bottom=y_flat - 0.05, top=y_flat + 0.05))

# Create a new plot
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools="", toolbar_location=None)

# Add a quad glyph for the heatmap
# Refer to columns from the ColumnDataSource
p.quad(left='left', right='right', bottom='bottom', top='top',
       fill_color=transform('data', color_mapper), line_color=None, source=source)

# Add a color bar to the right of the plot
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=12, location=(0,0), title="Value")
p.add_layout(color_bar, 'right')

# Show the results
show(p)
```
